## Coulomb's Law & Electrostatic Energy

Coulomb's Law is a fundamental principle in physics that describes the electrostatic force between two charged particles. The law states that the force between two point charges is directly proportional to the product of their charges and inversely proportional to the square of the distance between them. Mathematically, it can be expressed as:

$$
F = k \frac{q_1  q_2}{r^2}
$$

Where:
- $F$ is the electrostatic force between the charges,
- $k$ is Coulomb's constant ( k = 8.9875 $\times 10^9$ N m$^2$C$^2$ ),
- $q_1$ and $q_2$ are the magnitudes of the charges, and
- $r$ is the distance between the charges.

Coulomb's Law describes the behavior of electric charges, indicating that like charges repel each other, while opposite charges attract.

Electrostatic energy, also known as electric potential energy, is the potential energy stored in a system of charged particles due to their positions and interactions. Coulomb's Law is fundamental in understanding electrostatic energy.

On the other hand, this electrostatic force can do work if one of the charges is moved against the force. The work done in moving the charges against the electrostatic force is stored as electrostatic potential energy in the system.

The formula for the electrostatic potential energy ($U$) between two point charges ($q_1$ and $q_2$) separated by a distance ($r$) is derived from Coulomb's Law. It is given by:

$$
U = k\frac{ q_1 q_2}{r}
$$

This formula shows that the electrostatic potential energy between two charges is directly proportional to the product of the magnitudes of the charges and inversely proportional to the distance between them.

If the charges are of the same sign (both positive or both negative), the potential energy is positive, indicating that work must be done to bring them together (since like charges repel). If the charges are of opposite signs, the potential energy is negative, indicating that work is done by the electrostatic force as they come together (since opposite charges attract).


<h3>Task 1</h3>

Create a grid for $r$ with $h=0.1$ and $n=1000$. Calculate electrostatic force between $q_1=2$ C and $q_2=3$ C on the grid. Plot the force.


```python
import matplotlib.pyplot as mpl_p
h = 0.1
n = 1000
q_1, q_2 = 2, 3
k = 9e9

#
# your code here
#
grid = [i * h for i in range(1, n)]


func = [k * q_1 * q_2 / r ** 2 for r in grid]

mpl_p.plot(grid, func)
mpl_p.show()
```

## Hubble's Law

Hubble's law is a statement of a direct correlation between the distance to a galaxy and its recessional velocity as determined by the red shift. It can be stated as

$$
v = H_0 r,
$$

where $v$ is recessional velocity, $r$ is distance and $H_0$ is Hubble parameter.

$$
H_0 = 2.3\times 10^{-18} \text{ s}^{-1}
$$

<h3>Task 3</h3>

Create a square grid for $x$ and $y$ where

$$
r = \sqrt{x^2+y^2}.
$$

Take the grid step as $h=0.1$ and consider the limits of each direction from -100 to 100. Then, calculate recessional velocity on the grid and plot it with a heatmap using matplotlib module.

```python
import matplotlib.pyplot as mpl_p
h = 0.1

from math import sqrt
#
# your code here
#
x_min, x_max = -100, 100
n_x = int((x_max - x_min) / h) + 1
x_grid = [i * h + x_min for i in range(n_x)] ######### +x_min yazmadık - (-x_min yazdık)

y_min, y_max = -100, 100
n_y = int((y_max - y_min) / h) + 1
y_grid = [i * h + y_min for i in range(n_y)]

H_0 = 2.3e-18

velocities = []
for y in y_grid:
    velocities.append([])
    for x in x_grid:
        r = sqrt(x ** 2 + y ** 2)
        vel = H_0 * r
        velocities[-1].append(vel)

fig, ax = mpl_p.subplots() #############bu kısmı kopyaladık matplotlib den. Ayrıca hoca bunları finalde kendisi verecek eğer soracak olursa.
im = ax.imshow(velocities)
mpl_p.show
```
