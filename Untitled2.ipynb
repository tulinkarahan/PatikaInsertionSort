- **Equation of Motion**: The rate of change of the concentration of a chemical species $A$ in a reaction can be described by a differential equation of the form:
$$
\frac{d[A]}{dt} = -k[A]^n
$$
where $k$ is the rate constant and $n$ is the order of the reaction. This equaion is which is the same as $y'=f(x,y)$.

```python
f = lambda A: - k * A ** order

order = 1.0
k = 1.0

t_0 = 0.0
t_max = 10.0
dt = 0.1
n = int((t_max - t_0) / dt)

A_0 = 1.0

t_values = [t_0]
A_values = [A_0]
for i in range(n):
    t_values.append(t_values[i] + dt)
    A_values.append(A_values[i] + dt * f(A_values[i]))

import matplotlib.pyplot as mpl_p

mpl_p.plot(t_values, A_values)
mpl_p.xlabel("Time (sec)")
mpl_p.ylabel("Concentration (%)")
#mpl_p.xlim(4.0, 8.0)
#mpl_p.ylim(-0.05, 0.02)
mpl_p.grid()
mpl_p.show()

```

```python
def chemical_kinetics(A_0=1.0, order=1.0, k=1.0, t_max=5.0, dt=0.1):
    f = lambda A: - k * A ** order
    t_0 = 0.0
    n = int((t_max - t_0) / dt)

    t_values = [t_0]
    A_values = [A_0]
    for i in range(n):
        t_values.append(t_values[i] + dt)
        A_values.append(A_values[i] + dt * f(A_values[i]))

    return t_values, A_values

import matplotlib.pyplot as mpl_p

t_values, A_values = chemical_kinetics()

mpl_p.plot(t_values, A_values)
mpl_p.xlabel("Time (sec)")
mpl_p.ylabel("Concentration (%)")
#mpl_p.xlim(4.0, 8.0)
#mpl_p.ylim(-0.05, 0.02)
mpl_p.grid()
mpl_p.show()
```
