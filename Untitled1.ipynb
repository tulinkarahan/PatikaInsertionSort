- **Equations of Motion**:
$$
      \frac{d^2r}{dt^2} = -\frac{G M_1 M_2}{r^2} \left( 1 + \frac{3}{2} \left( \frac{v}{c} \right)^2 \right)
$$
where $v$ is the relative velocity of the stars. If the stars are in circular orbit, then,
$$
v = 2\pi\frac{r}{T}.
$$
Here $r$ is the distance and $T$ is the orbital period.
- Model the emission of gravitational waves from a binary star system. Solve the equations of motion for the orbit and analyze the resulting gravitational waveforms.


$$
      \frac{d^2r}{dt^2} = -\frac{G M_1 M_2}{r^2} \left( 1 + \frac{3}{2} \left( \frac{v}{c} \right)^2 \right)\\
      \frac{dv}{dt} = -\frac{G M_1 M_2}{r^2} \left( 1 + \frac{3}{2} \left( \frac{v}{c} \right)^2  \right)\\
      \frac{dr}{dt} = v = 2 \pi \frac{r}{T}
$$


```python
import matplotlib.pyplot as mpl_p
from math import pi

f = lambda v, r: - G * M_1 * M_2 * (1 + 1.5 * (v / c) ** 2) / r ** 2

G, c = 1., 1.
T, M_1, M_2 = 1.0, 2.2224 * 2e-1, 17.80 * 2e-1
t_0, r_0 = 0.0, c
v_0 = 2 * pi * r_0 / T
print(v_0)

dt = 0.1
t_max = 1e1
n = int((t_max - t_0) // dt)
print(n)

t_list = [t_0]
v_list = [v_0]
r_list = [r_0]
for i in range(n):
    print(i)
    elem = t_list[i] + dt
    t_list.append(elem)
    print("t:", elem)

    elem = v_list[i] + dt * f(v_list[i], r_list[i])
    v_list.append(elem)
    print("v:", elem)

    elem = r_list[i] + dt * 2 * pi * r_list[i] / T
    r_list.append(elem)
    print("r:", elem)
    print()
    print()

mpl_p.plot(t_list, r_list, label="Distance")
mpl_p.legend()
mpl_p.show()
mpl_p.plot(t_list, v_list, label="Velocity")
mpl_p.legend()
mpl_p.show()
```
